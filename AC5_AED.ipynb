```python
import pandas as pd 
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import plotly.express as px

dataframe = pd.read_csv('CarPrice_Assignment.csv')
price = dataframe['price']
lista = ['curbweight', 'wheelbase', 'enginesize', 'horsepower', 'compressionratio']

tabela = {
    'Par de Variável':[],
    'Pearsons correlation':[]
          }
          
for i in lista:
  corr, _ = pearsonr(price, dataframe[i])
  tabela['Par de Variável'].append(i)
  tabela['Pearsons correlation'].append(corr)
```

```python
fig = px.scatter(dataframe, x = "price", y = "curbweight", trendline = 'ols', trendline_color_override='red', width = 800)
corr, _ = pearsonr(price, dataframe['curbweight'])
fig.show()
print(f'Pearsons correlation: {corr:.4f}')
```
